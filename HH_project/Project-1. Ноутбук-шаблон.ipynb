```python
#ваш код здесь
def experience(arg):
    if (str(arg) != 'nan') and (arg != 'Не указано'):
        experience_in_month = 0
        lst = arg.split()
        if lst[3][:3] == 'год' or lst[3][:3] == 'лет':
            experience_in_month += int(lst[2])*12
        if lst[3][:3] == 'мес':
            experience_in_month += int(lst[2])
        if lst[5][:3] == 'мес':
            experience_in_month += int(lst[4])
        return experience_in_month

hh_copy["Опыт работы (месяц)"] = hh_copy['Опыт работы'].apply(experience)
print (hh_copy["Опыт работы (месяц)"].median())
hh_copy = hh_copy.drop(columns='Опыт работы')
hh_copy.info()
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

def check_town(arg):
    tmp = arg.strip()
    if tmp == "Москва" or  tmp == "Санкт-Петербург": return  tmp
    for s in million_cities:
        if tmp == s: return 'город-миллионник'
    return 'другие'

def readiness(arg):
    if arg is None : return False
    return arg.find('не') == -1


hh_lst = list(hh_copy['Город, переезд, командировки'])
lst=[]
for s in hh_lst:
    tmp = (s.split(' ,'))
    sample = [ None,  None,  None]
    sample[0] = tmp[0].partition('(')[0]
    for i in range(1,len(tmp)):
        if tmp[i].find('пере') != -1: sample[1] = tmp[i].partition('(')[0]    
        if tmp[i].find('ком') != -1: sample[2] = tmp[i].partition('(')[0]
    lst.append(sample)

tmp_df = pd.DataFrame(data = lst, columns= ['town', 'переезд', 'командировки'])

tmp_df['Город'] = tmp_df['town'].apply(check_town)
tmp_df['Готовность к переезду'] = tmp_df['переезд'].apply(readiness)
tmp_df['Готовность к командировкам'] = tmp_df['командировки'].apply(readiness)

print (round(tmp_df[tmp_df['Город'] == 'Санкт-Петербург'].shape[0]/tmp_df.shape[0]*100))
print (round(tmp_df[tmp_df['Готовность к переезду'] & tmp_df['Готовность к командировкам']] .shape[0]/tmp_df.shape[0]*100))

tmp_df = tmp_df.drop(columns=['town', 'переезд', 'командировки'])
hh_copy = hh_copy.join(tmp_df)
hh_copy = hh_copy.drop(columns=['Город, переезд, командировки'])

hh_copy.info()
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
type_occupation = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
type_schedule = ['полный день', 'сменный график', 'гибкий график', 'удалённая работа', 'вахтовый метод']

def occupation_and_schedule(arg,col):
    lst = arg.split(', ')
    for s in lst:
        if s == col: return True
    return False

for occupation in type_occupation:
    hh_copy[occupation] = hh_copy['Занятость'].apply(lambda arg: occupation_and_schedule(arg,occupation))

for schedule in type_schedule:
    hh_copy[schedule] = hh_copy['График'].apply(lambda arg: occupation_and_schedule(arg,schedule))

print (hh_copy[hh_copy['проектная работа'] & hh_copy['волонтерство']].shape[0])
print (hh_copy[hh_copy['вахтовый метод'] & hh_copy['гибкий график']].shape[0])

hh_copy = hh_copy.drop(columns=['Занятость', 'График'])

hh_copy.info()
```

```python
currency_dict ={'грн.':	['UAH',	10],'USD':	['USD',	1],'EUR':	['EUR',	1],'бел.руб.':	['BYN',	1],'KGS':	['KGS',	10],'сум':	['UZS',	10000],'AZN':	['AZN',	1],'KZT':	['KZT',	100],'руб.':	['RUB',	1]}
tmp_df['ЗП'] = hh_copy['ЗП']

tmp_df['sum'] = tmp_df['ЗП'].apply(lambda arg: float(arg.split()[0]))
tmp_df['currency txt'] = tmp_df['ЗП'].apply(lambda arg: arg.split()[1]) 
tmp_df['currency'] = tmp_df['currency txt'].apply(lambda arg: currency_dict[arg][0]) 
tmp_df['date']= hh_copy['Обновление резюме']

union_df = tmp_df.merge(df_currency, on = ['date','currency'], how = 'left')
union_df['close'] = union_df.apply(lambda df: 1 if df['currency'] == 'RUB' else df['close'],axis=1)
union_df['proportion'] = union_df.apply(lambda df: 1 if df['currency'] == 'RUB' else df['proportion'],axis=1)
union_df['ЗП (руб)'] = union_df['sum'] * union_df['close'] / union_df['proportion']  

#display(union_df.head())
#display(union_df.info())

hh_copy['ЗП (руб)']  = union_df['ЗП (руб)'] 
display(hh_copy.info())

print (hh_copy['ЗП (руб)'].median())
```
